```python
import numpy as np
import pandas as pd

data = np.random.rand(10, 3) * 100
df = pd.DataFrame(data, columns=['A', 'B', 'C'])

nan_indices = np.random.choice(df.index, size=3, replace=False) 
df.loc[nan_indices, 'A'] = np.nan  
df.loc[nan_indices, 'B'] = np.nan

print(df)

df.fillna(df.mean(), inplace=True)
print(df)

df_sorted = df.sort_values(by=['A', 'B'])
print(df_sorted)
```

```python
import pandas as pd
data1 = {
    'Student': [1, 2, 3, 4],
    'Name': ['A', 'B', 'C', 'D'],
    'Roll no': [101, 102, 103, 104]
}
df1 = pd.DataFrame(data1)
data2 = {
    'Roll no': [101, 102, 103, 104],
    'Subject': ['A105', 'Q104', 'D102', 'S104']
}
df2 = pd.DataFrame(data2)
merged_df = pd.merge(df1, df2, on='Roll no')
print(merged_df)

```
