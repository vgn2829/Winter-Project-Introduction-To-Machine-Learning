### Q1.
Using NumPy, create a 3D array of shape (4, 4, 4) filled with random numbers between 1 and 100. Find the z score normalisation of each element of the array.

```python
#1

val = np.random.randint(1, 101, size=(4, 4, 4))
mean = np.mean(val)
std_dev = np.std(val)
z_score_normalized = (val - mean) / std_dev
```

### Q5.
Using NumPy, calculate the dot product of two matrices and find the determinant of the resulting matrix.

```python
a = np.array([
    [1, 2, 3],
    [1, 2, 3],
    [1, 2, 3]
])

b = np.array([
    [1, 0, 0],
    [0, 1, 0],
    [0, 0, 1]
])
```

```python
a @ b
print(f"determinant{np.linalg.det(a@b)}")
```

### Q6.
With Pandas, group a DataFrame by a categorical column and calculate the sum of another numerical column for each group.

```python
data = {
    'Category': ['A', 'B', 'A', 'B', 'C', 'A', 'C'],
    'Values': [10, 20, 30, 40, 50, 60, 70]
}

df = pd.DataFrame(data)

group = df.groupby('Category')['Values'].sum()
print(group)
```

### Q8.
Using NumPy, create a 2D array and perform a slicing operation to extract a specific sub-array.

```python
a = np.array([
    [0, 0],
    [1, 1],
    [2, 2],
    [3, 3]
])

print(a[:1]) # takes out first
print(a[1:]) # takes out all leaving first
print(a[1:len(a)//2]) # takes out middle
```

### Q9.
 Using Pandas, perform a merge operation between two DataFrames based on a common column for dataset of your choice.

```python
data1 = {
    'Category': ['A', 'B'],
    'Values': [10, 20]
}

data2 = {
    'Category': ['C', 'B'],
    'Values': [30, 40]
}

df1 = pd.DataFrame(data1)
df2 = pd.DataFrame(data2)

merge = pd.merge(df1, df2, on='Category', how='outer')
merge
```
