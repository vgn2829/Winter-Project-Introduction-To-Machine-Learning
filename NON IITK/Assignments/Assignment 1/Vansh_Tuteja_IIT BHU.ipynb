```python
print("3rd Question ")
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 10, 100)
y = np.sin(x)

plt.plot(x, y, label='sin(x)', color='b', linestyle='-', linewidth=2)
plt.title('Trend of sin(x)')
plt.xlabel('X Axis (Time)')
plt.ylabel('Y Axis (Amplitude)')
plt.legend()
plt.show()
```

```python
print("4th Question ")
import pandas as pd
import numpy as np

np.random.seed(42)
random_data = {
    "Scores_A": np.random.randn(10),
    "Scores_B": np.random.randn(10),
    "Scores_C": np.random.randn(10)
}
grades_df = pd.DataFrame(random_data)

grades_df.loc[2, "Scores_A"] = np.nan
grades_df.loc[7, "Scores_B"] = np.nan

print("Original DataFrame with NaN values:")
print(grades_df)

grades_df.fillna(grades_df.mean(), inplace=True)

print("\nDataFrame after replacing NaN values with the mean:")
print(grades_df)

sorted_grades = grades_df.sort_values(by=["Scores_A", "Scores_B"], ascending=[True, False])

print("\nDataFrame sorted by Scores_A (ascending) and Scores_B (descending):")
print(sorted_grades)
```

```python
print("6th Question ")
import pandas as pd

data = {
    "Category": ["A", "B", "A", "C", "B", "C", "A", "B", "C", "A"],
    "Values": [10, 20, 15, 40, 25, 35, 10, 20, 30, 15]
}
df = pd.DataFrame(data)

grouped_sum = df.groupby("Category")["Values"].sum()

print("Sum of 'Values' for each 'Category':")
print(grouped_sum)
```

```python
print("9th Question ")
import pandas as pd

students = {
    "Student_ID": [1, 2, 3, 4],
    "Name": ["Vansh", "Rohan", "Suraj", "Manas"],
    "Age": [20, 21, 19, 22]
}
students_df = pd.DataFrame(students)

scores = {
    "Student_ID": [1, 2, 3, 5],
    "Math_Score": [85, 90, 78, 88],
    "Science_Score": [89, 92, 80, 85]
}
scores_df = pd.DataFrame(scores)

merged_df = pd.merge(students_df, scores_df, on="Student_ID", how="inner")

print("Merged DataFrame:")
print(merged_df)

```
